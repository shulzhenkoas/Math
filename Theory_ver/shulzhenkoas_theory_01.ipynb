#### Задание 1.
##### Из колоды в 52 карты извлекаются случайным образом 4 карты. Найти вероятность того, что все карты – крести. Найти вероятность, что среди этих карт окажется хотя бы один туз.

В колоде 13 крестовых карт. Карты вытаскиваем одну за другой, значит вероятности вытащить 4 крестовых карты подряд можно представить как 4 независимых события, которые должны произойти совместно:

Вероятность вытащить первую крестовую карту:
$$\frac{13}{52}$$

Вероятность вытащить вторую крестовую карту (так как одну крестовую карту уже вытащили, то их стало на одну меньше, так же как и карт в колоде):
$$\frac{12}{51}$$


Вероятность вытащить третью крестовую карту:
$$\frac{11}{50}$$

Вероятность вытащить четвертую крестовую карту:
$$\frac{10}{49}$$

Вероятность вытащить подряд 4 крести:
$$\frac{13}{52} * \frac{12}{51} * \frac{11}{50} * \frac{10}{49} = \frac{1}{17} * \frac{11}{5} * \frac{1}{49} = \frac{11}{4165}$$

Вероятность вытащить крестового туза из 13 крестовых карт:
$$\frac{1}{13}$$

Вероятность вытащить 4 крестовых карты, одна из которых туз:
$$\frac{1}{13} * \frac{11}{4165} = \frac{11}{54145}$$


#### Задание 2.
##### На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Вероятность подобрать код будет 1 деленая на количество возможных сочетаний 3 кнопок из 10 (порядок цифр не имеет значения, так как кнопки нажимаются одновременно).

Количество возможных сочетаний:
$$C_{10}^{3}=\frac{10!}{3! * (10 - 3)!}=\frac{7! * 8 * 9 * 10}{2 * 3 * 7!}=120$$

Вероятность подобрать код с первого раза:
$$\frac{1}{120}$$

#### Задание 3.
##### В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

По аналогии с первой задачей получаем вероятность, что все три извлеченные детали окрашены будет:
$$\frac{9}{15} * \frac{8}{14} * \frac{7}{13} = \frac{3}{5} * \frac{4}{1} * \frac{1}{13} = \frac{12}{65}$$

#### Задание 4.
##### В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

По аналогии с первой задачей получаем вероятность, что оба билеты будут выигрышными:
$$\frac{2}{100} * \frac{1}{99} = \frac{1}{50} * \frac{1}{99} = \frac{1}{4950}$$
