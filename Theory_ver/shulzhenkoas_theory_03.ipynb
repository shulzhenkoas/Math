```python
import numpy as np
from math import factorial
```

$$C_n^k = \frac{n!}{n!(n-k)!}$$
$$P_n(X=k) = C_n^k p^k q^{n-k}$$
$$P_m \approx \frac{\lambda^m}{m!}e^{-\lambda}$$

```python
def C(k, n):
    return factorial(n) / (factorial(k) * factorial(n - k))
```

```python
def bernulli(k, n, p):
    return C(k, n) * (p**k) * ((1 - p)**(n - k))
```

```python
def poisson(m, n, p):
    lambda_ = n * p
    return lambda_**p / factorial(m) * np.exp(-lambda_)
```

Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
zp = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
zp.shape
```

```python
mean_zp = zp.sum() / zp.size
print("Среднее арифметическое:", mean_zp)
print("Проверка через метод numpy:", zp.mean())
```

```python
std_zp = np.sqrt(((zp - mean_zp)**2).sum() / zp.size)
print("Среднее квадратичное отклонение:", std_zp)
print("Среднее квадратичное отклонение через метод numpy:", zp.std(ddof=0))
```

```python
variance1_zp = (((zp - mean_zp)**2).sum() / zp.size)
print("Смещенная дисперсия:", variance1_zp)
print("Смещеная дисперсия через метод numpy:", zp.var(ddof=0))
```

```python
variance2_zp = (((zp - mean_zp)**2).sum() / (zp.size - 1))
print("Несмещенная дисперсия:", variance2_zp)
print("Несмещеная дисперсия через метод numpy:", zp.var(ddof=1))
```

#### Задание 3.
##### На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

Воспользуемся формулой Байеса:
$$P(B|A) = \frac {P(B) \cdot P(A|B)}{P(A)}$$
Событием A будем считать попадание в мишень, а событиями B1, B2 и B3 — что выстрел совершил первый, второй и третий спортсмены:
$$P(A|B1) = 0.9$$
$$P(A|B2) = 0.8$$
$$P(A|B3) = 0.6$$
Вероятности событий B1, B2 и B3 — что выстрел совершил первый, второй и третий спортсмены:
$$P(B1) = P(B2) = P(B3) = \frac{1}{3}$$
Вероятность события A — попадание в мишень равна:
$$P(A) = \frac{1}{3} \cdot 0.9 + \frac{1}{3} \cdot 0.8 + \frac{1}{3} \cdot 0.6$$

```python
pb1 = pb2 = pb3 = 1/3
pab1 = 0.9
pab2 = 0.8
pab3 = 0.6
pa1 = pb1 * pab1 + pb2 * pab2 + pb3 * pab3
pa1
```

```python
print("Вероятность того, что стрелял первый спортсмен:", pb1 * pab1 / pa1)
print("Вероятность того, что стрелял второй спортсмен:", pb2 * pab2 / pa1)
print("Вероятность того, что стрелял третий спортсмен:", pb3 * pab3 / pa1)
```

##### В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

Воспользуемся формулой Байеса:
$$P(B|A) = \frac {P(B) \cdot P(A|B)}{P(A)}$$
Событием A будем считать сдачу студентом сессию, а событиями B1, B2 и B3 — что студент с факультета "A", "B" и "C":
$$P(A|B1) = 0.8$$
$$P(A|B2) = 0.7$$
$$P(A|B3) = 0.9$$
Вероятности событий B1, B2 и B3 — что студент с факультета "A", "B" и "C":
$$P(B1) = P(B2) = \frac{1}{4}$$
$$P(B3) = \frac{1}{2}$$
Вероятность события A — сдача сессии студентом равна:
$$P(A) = \frac{1}{4} \cdot 0.8 + \frac{1}{4} \cdot 0.7 + \frac{1}{2} \cdot 0.9$$

```python
pd1 = pd2 = 1/4
pd3 = 1/2
pad1 = 0.8
pad2 = 0.7
pad3 = 0.9
pa2 = pd1 * pad1 + pd2 * pad2 + pd3 * pad3
pa2
```

```python
print('Вероятность того, что студент сдавший сессию учится на факультете "A":', pd1 * pad1 / pa2)
print('Вероятность того, что студент сдавший сессию учится на факультете "B":', pd2 * pad2 / pa2)
print('Вероятность того, что студент сдавший сессию учится на факультете "C":', pd3 * pad3 / pa2)
```
