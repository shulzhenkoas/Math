#### 1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import math
import matplotlib.pyplot as plt

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.linspace(-10, 10, 100)
k1, a1, b1 = 7, 1, 5
k2, a2, b2 = 0.5, 5, 3
k3, a3, b3 = -3, -5, -7
plt.plot(x, k1 * np.cos(x - a1) + b1, label='k1=7; a1=1; b1=5')
plt.plot(x, k2 * np.cos(x - a2) + b2, label='k2=0.5; a2=5; b2=3')
plt.plot(x, k3 * np.cos(x - a3) + b3, label='k3=-3; a3=-5; b3=-7')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

#### 4. Задание (в программе)
1) Решите систему уравнений:
##### y = x2 – 1
##### exp(x) + x*(1 – y) = 1

```python
from scipy.optimize import fsolve

def equatuins(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x *(1 - y) - 1)

x1, y1 = fsolve(equatuins, (1, 1))
x2, y2 = fsolve(equatuins, (5, 5))
print("x1 =", x1, "y1 =", y1)
print("x2 =", x2, "y2 =", y2)
```
